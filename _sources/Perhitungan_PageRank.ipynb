# PageRank

PageRank adalah algoritma yang digunakan oleh mesin pencari, terutama oleh Google, untuk menilai pentingnya suatu halaman web. Algoritma ini dinamai dari salah satu pendiri Google, Larry Page. PageRank telah menjadi bagian integral dari algoritma pencarian Google dan masih digunakan sebagai salah satu faktor dalam menentukan peringkat halaman web dalam hasil pencarian. Namun, seiring waktu, Google telah mengembangkan dan memodifikasi berbagai aspek algoritma pencariannya untuk meningkatkan relevansi dan kualitas hasil pencarian. Tujuan utama PageRank adalah memberikan peringkat kepada halaman-halaman web berdasarkan seberapa penting dan otoritatifnya dalam ekosistem tautan web.
Proses perhitungan PageRank melibatkan matriks tautan dari seluruh web, yang mencerminkan hubungan tautan antara halaman-halaman web. Algoritma kemudian melakukan iterasi untuk menghitung nilai PageRank untuk setiap halaman berdasarkan matriks tautan dan mengupdate nilai-nilai ini hingga konvergensi.Rumus PageRank sebagai berikut:




$$
PR(u) = (1-d) . \frac{1}{N} + d . \sum_{v∈Bu} \frac{PR(v)}{L(v)}
$$

* PR(u) : PageRank dari halaman (u)
* d : damping factor, biasanya diatur 0,85
* N : jumlah total halaman dalam graf (jumlah node dalam graf)
* B_u : himpunan halaman yang memiliki tautan ke halaman (u)
* L(v) : jumlah tautan keluar dari halaman (v)
